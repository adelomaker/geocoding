```python
#Importo le librerie necessarie
import pandas as pd
import numpy as np
import urllib
from urllib.parse import quote
import requests
import folium
```

#### Per ricavare l'elenco delle Istituzioni Scolastiche Pugliesi, utilizzo gli Open Data dell'USR Puglia disponibili al seguente indirizzo https://www.pugliausr.gov.it/index.php/70-la-scuola-in-numeri

```python
#Carico i necessari dati
data = None
data = pd.read_excel("./DATA/rete_scolastica_Puglia_ 2020-21.xlsx", sheet_name="Sede_Direttivo",engine='openpyxl')
data
```

```python
#Creo un nuovo DataFrame con i dati di interesse
data_lecce = data[(data['Provincia'] == 'Lecce')].reset_index(drop=True)
data_lecce
```

```python
#Stampo a video per verificare la tipologia di informazione
print('Indirizzo: ' + data_lecce['Indirizzo'][0] + '\n\nCOMUNE: ' + data_lecce['Comune'][0])
```

```python
#Verifica modalità di funzionamento parse.quote
urllib.parse.quote(data_lecce['Indirizzo'][0] + data_lecce['Comune'][0])
```

```python
#Estraggo il valore di lat e lon
txt1=data_lecce['Indirizzo'][0]
txt2=data_lecce['Comune'][0]
url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(txt1 +' '+ txt2) +'?format=json'
response = requests.get(url).json()
print('Latitude: '+response[0]['lat']+', Longitude: '+response[0]['lon'])
```

```python
#Creo la funzione geocode per estrarre i valori di lat e lon di tutto il data
def geocode(indirizzo, comune):
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(indirizzo + ' ' + comune) +'?format=json'
    response = requests.get(url).json()
    if(len(response)!=0):
        return(response[0]['lat'], response[0]['lon'])
    else:
        return('-1')

#Applico la funzione ad ogni record presente nel DataFrame
data_lecce["Coordinate"]=None
coordinate=[]
for i in range(0,len(data_lecce)):
    indirizzo= data_lecce["Indirizzo"][i]
    comune = data_lecce['Comune'][i]
    risposta=geocode(indirizzo, comune)
    coordinate.append(risposta)
    
#Aggiungo nel Dataframe una colonna con le coordinate 
data_lecce['Coordinate']=coordinate

#Visualizzo i primi 5 record del DataFrame
data_lecce.head()
```

```python
#Salvo il Dataframe in un file .csv
data_lecce.to_csv('./DATA/Scuola_Lecce.csv',index=False)
```
