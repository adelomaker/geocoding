```python
#Importo le librerie necessarie
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
```

```python
#Carico i dati
data = None
data_file = './DATA/Scuola_Lecce.csv'
with open(data_file,'r') as f:
    data = pd.read_csv(f)
    
data
```

```python
# Individuo i valori medi per centrare la mappa
data['latitudine'] = data['latitudine'].astype('float')
data['longitudine'] = data['longitudine'].astype('float')
lat_M = data['latitudine'].mean()
lon_M = data['longitudine'].mean()
```

```python
#Creo un oggetto mappa base
m=folium.Map(location=[lat_M, lon_M], zoom_starter=10)
```

```python
m
```

```python
markercluster = MarkerCluster().add_to(m)
```

```python
#Traccio gli indicatori sulla mappa

#Tooltip
tooltip = "Fare click per informazioni"

i=0
for i in range(0,len(data)):
    codice=data.at[i,'Codice']
    denominazione=data.at[i,'Denominazione']
    dirigente=data.at[i,'Dirigente']
    folium.Marker([data['latitudine'][i], data['longitudine'][i]],
            popup= f"Codice: {codice} Denominazione: {denominazione} Dirigente: {dirigente}",
            tooltip=tooltip,
            icon=folium.Icon(prefix='fa',icon='graduation-cap')).add_to(markercluster)
```
